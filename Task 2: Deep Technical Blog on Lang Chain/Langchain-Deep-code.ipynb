# 2. Import Libraries

```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import PromptTemplate
```

# 3. Load Offline Model (Lightweight)

```python
pipe = pipeline(
    "text-generation",
    model="distilgpt2",   # lightweight model
    max_length=100,
    temperature=0.7
)
```

# 4. Wrap with LangChain & Create Prompt + Chain (LCEL)

```python
llm = HuggingFacePipeline(pipeline=pipe)
prompt = PromptTemplate.from_template("Explain {topic} in simple terms")

chain = prompt | llm

response = chain.invoke({"topic": "Artificial Intelligence"})

print(response)
```

```python
# Install (if needed)
!pip install langchain-classic

# Import
from langchain_classic.memory import ConversationBufferMemory

# Create memory
memory = ConversationBufferMemory()

# Store conversation
memory.save_context({"input": "Hi"}, {"output": "Hello!"})
memory.save_context({"input": "What is AI?"}, {"output": "AI is Artificial Intelligence"})

# Retrieve memory
print(memory.load_memory_variables({}))
```

# Simple Agent

```python
def simple_agent(query):
    if "math" in query.lower():
        return "This looks like a math problem."
    elif "ai" in query.lower():
        return "This is related to Artificial Intelligence."
    else:
        return "General query."

print(simple_agent("Tell me about AI"))
```

# Mini Use Case

```python
# Sample knowledge base
documents = [
    "LangChain is used to build LLM applications",
    "AI stands for Artificial Intelligence",
    "Machine Learning is a subset of AI"
]

# Simple retrieval
def retrieve(query):
    for doc in documents:
        if query.lower() in doc.lower():
            return doc
    return "No relevant info found"

# Ask question
query = "What is AI?"
context = retrieve(query)

# Generate answer
response = chain.invoke({"topic": context})

print(response)
```
