**Task 1: PromptTemplate (No Hardcoding)**

```python
# Import PromptTemplate from langchain_core (latest version)
from langchain_core.prompts import PromptTemplate

# Create a PromptTemplate object
# This template takes one input variable: "topic"
# It defines a reusable prompt structure instead of hardcoding
template_1 = PromptTemplate(
    input_variables=["topic"],  # List of variables to be dynamically filled
    template="Explain {topic} in simple terms for beginners"  # Template with placeholder
)

# Format the template by passing a value for "topic"
# This replaces {topic} with "Artificial Intelligence"
output = template_1.format(topic="Artificial Intelligence")

# Print the generated prompt
# This shows the final structured prompt after variable substitution
print(output)
```

**Task 2: Multi-Input Prompt**

```python
# Import PromptTemplate (latest LangChain version)
from langchain_core.prompts import PromptTemplate

# Create a PromptTemplate with multiple input variables
# This template accepts: topic, audience, and tone
template_2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],  # Dynamic inputs
    template="Explain {topic} for {audience} in a {tone} tone"  # Structured prompt
)

# Test Case 1
# Replaces placeholders with given values
# topic = AI, audience = beginners, tone = friendly
print(template_2.format(topic="AI", audience="beginners", tone="friendly"))

# Test Case 2
# Demonstrates reusability of the same template with different inputs
print(template_2.format(topic="Python", audience="kids", tone="fun"))

# Test Case 3
# Shows flexibility for a more technical audience and tone
print(template_2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

**Task 4: ChatPromptTemplate**

```python
# Import ChatPromptTemplate from langchain_core
from langchain_core.prompts import ChatPromptTemplate


# Function to generate chat-based prompts using roles
def get_chat_prompt(role, topic):

    # Define behavior for each role
    # This acts as the "system message" guiding the AI's behavior
    role_behaviors = {
        "teacher": "You are a helpful teacher explaining concepts clearly.",
        "interviewer": "You are an interviewer asking challenging questions.",
        "motivator": "You are a motivational speaker encouraging learning."
    }

    # Create a ChatPromptTemplate using tuple-based message format
    # ("system", ...) defines behavior
    # ("human", ...) represents user input
    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", role_behaviors.get(role, "You are a helpful assistant.")),
        ("human", "Explain {topic}")
    ])

    # Format messages by replacing {topic} with actual input
    # Returns structured chat messages (not plain text)
    return chat_prompt.format_messages(topic=topic)


# Test Case
# -------------------------------

# Generate chat messages for role = teacher and topic = Neural Networks
messages = get_chat_prompt("teacher", "Neural Networks")

# Print each message in the chat structure
for msg in messages:
    print(msg)
```

**Task 5: Input Validation**

```python
# Function to validate user inputs for audience and tone
def validate_inputs(audience, tone):

    # Define allowed values for audience and tone
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Check if the provided audience is valid
    # If not, assign a default value
    if audience not in valid_audience:
        print("Invalid audience! Default → beginner")
        audience = "beginner"

    # Check if the provided tone is valid
    # If not, assign a default value
    if tone not in valid_tone:
        print("Invalid tone! Default → formal")
        tone = "formal"

    # Return validated (or corrected) values
    return audience, tone

# Test Case
# -------------------------------

# Passing invalid inputs to test validation logic
print(validate_inputs("kids", "funny"))
```

**Task 6: Prompt Generator**

```python
# Import PromptTemplate
from langchain_core.prompts import PromptTemplate

# Function to generate a dynamic prompt based on inputs
def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs (audience and tone)
    # Ensures only allowed values are used
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Define different prompt styles
    # Each style represents a different way of framing the prompt
    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    # Step 3: Create PromptTemplate dynamically based on selected style
    # style determines which template is used
    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=style_templates.get(style)  # Fetch selected style template
    )

    # Step 4: Format the template with actual input values
    # Replaces placeholders with user inputs
    return template.format(topic=topic, audience=audience, tone=tone)


# Test Case
# -------------------------------

# Generate a prompt using storytelling style
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7: Reusability Test**

```python
# Import PromptTemplate
from langchain_core.prompts import PromptTemplate

# Create a reusable PromptTemplate
# This single template will be used with multiple inputs
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],  # Dynamic inputs
    template="Explain {topic} for {audience} in a {tone} tone"  # Common structure
)

# Define multiple test cases
# Each tuple contains (topic, audience, tone)
inputs = [
    ("AI", "beginner", "formal"),
    ("Python", "intermediate", "casual"),
    ("Data Science", "expert", "formal"),
    ("Deep Learning", "beginner", "fun"),
    ("NLP", "intermediate", "casual")
]

# Loop through each input set
# Apply the same template to generate different prompts
for topic, audience, tone in inputs:

    # Format template with current values
    output = reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    # Print generated prompt
    print(output)
```

**final pipeline**

```python
def pipeline(topic, audience, tone, style):
    return generate_prompt(topic, audience, tone, style)

print(pipeline("AI", "beginner", "fun", "teaching"))
```
